# Text Data Analysis (Simple)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Load Data

```python
df = pd.read_csv("alice.csv")
```

## 2. Convert Text to Words

```python
df["Words"] = df["Text"].str.lower().str.split()
```

## 3. Count Words

```python
df["WordCount"] = df["Words"].apply(len)
```

## 4. Group Analysis

```python
total_result = df.groupby("Book")["WordCount"].sum()
mean_result = df.groupby("Book")["WordCount"].mean()
max_result = df.groupby("Book")["WordCount"].max()
min_result = df.groupby("Book")["WordCount"].min()
```

## 5. Visualization

```python
total_result.plot(kind='bar')
plt.title('Mean Word Count')
plt.show()
```

```python
mean_result.plot(kind='bar')
plt.title('Mean Word Count')
plt.show()
```

```python
max_result.plot(kind='bar')
plt.title('Mean Word Count')
plt.show()
```
