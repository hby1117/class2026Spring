### Sequence: String
```python
a = "Hello"
b = "World!"
```

### Sequence: List
```python
a = [1, 3, 5, 7, 9]
```
```python
# Access items
a[0]
a[1:3]
```
```python
# Modify items
a[0] = 11

# Append items
a.append(6)

# Remove items
a.pop(2)
```

### Dictionary
```python
a = {"name": "John", "age": 20}
```
```python
# Access values
a["name"]

# Modify values
a["name"] = "Mark"

# Add values
a["city"] = "Seoul"

# Remove values
del a["city"]

---

## 4. Exercises

1
<br />
<br />
Three students, A, B, and C, scored 85, 92, and 76 on their math exam, respectively.
<br />
What is the average score?

2
<br />
<br />
Combine these two names to create a fullname.
<br />
<br />
First Name: William
<br />
Last Name: Shakespeare

3
<br />
<br />
You recorded the number of steps you walked each day for a week.
<br />
```python
steps = [5230, 6100, 4890, 7200, 8100, 6500, 9000]
```
