# Pandas Practice

```python
import pandas as pd

dfw = pd.read_csv("weather.csv")
dfp = pd.read_csv("product.csv")
dfm = pd.read_csv("movies.csv")
```

## Part 1. Weather

1. Temperature column

```python
dfw["Temperature"]
```

2. Humidity와 City column

```python
dfw[["Humidity", "City"]]
```

3. 0번 row

```python
dfw.loc[0]
```

4. 1~3번 row

```python
dfw.loc[1:3]
```

5. Temperature이 21보다 높은 row

```python
cond = dfw["Temperature"] > 21
dfw[cond]
```

6. Humidity가 40 이상인 row

```python
cond = dfw["Humidity"] >= 40
dfw[cond]
```

7. Temperature가 21보다 크고 Humidity가 40보다 작은 row

```python
cond = (dfw["Temperature"] > 21) & (dfw["Humidity"] < 40)
dfw[cond]
```

## Part 2. Product

1. Price column

```python
dfp["Price"]
```

2. Category와 Item column

```python
dfp[["Category", "Item"]]
```

3. 0번 row

```python
dfp.loc[0]
```

4. 1번~4번 row

```python
dfp.loc[1:4]
```

5. Price가 1 이상인 row

```python
dfp[dfp["Price"] >= 1]
```

6. Quantity가 60 이상인 row

```python
cond = dfp["Quantity"] >= 60
dfp[cond]
```

7. Category가 "Fruit"이고 Quantity가 50보다 큰 row

```python
cond = (dfp["Category"] == "Fruit") & (dfp["Quantity"] > 50)
dfp[cond]
```

8. Category가 'Vegetable'이고 Price가 1보다 작은 row

```python
cond = (dfp["Category"] == "Vegetable") & (dfp["Price"] < 1)
dfp[cond]
```

9. Category별 평균 Price

```python
dfp.groupby("Category")["Price"].mean()
```

10. Category별 평균 Quantity

```python
dfp.groupby("Category")["Quantity"].mean()
```

11. Category별 최대 Quantity

```python
dfp.groupby("Category")["Quantity"].max()
```

12. Quantity가 평균보다 큰 row

```python
# dfp["Quantity"].mean()은 전체 row에서 Quantity의 평균입니다.

cond = dfp["Quantity"] > dfp["Quantity"].mean()
dfp[cond]
```

## Part 3. Movies

1. Title column

```python
dfm["Title"]
```

2. Genre와 Rating column

```python
dfm[["Genre", "Rating"]]
```

3. 2번 row

```python
dfm.loc[2]
```

4. 2~3번 row

```python
dfm.loc[2:3]
```

5. Rating이 8 이상인 row

```python
cond = dfm["Rating"] >= 8
dfm[cond]
```

6. Views가 180 이상인 row

```python
cond = dfm["Views"] >= 180
dfm[cond]
```

7. Genre가 'Sci-Fi'이고 Rating이 8보다 큰 row

```python
cond = (dfm["Genre"] == "Sci-Fi") & (dfm["Rating"] > 8)
dfm[cond]
```

8. Genre가 'Animation'이고 Views가 150보다 큰 row

```python
cond = (dfm["Genre"] == "Animation") & (dfm["Views"] > 150)
dfm[cond]
```

9. Genre별 평균 Rating

```python
dfm.groupby("Genre")["Rating"].mean()
```

10. Genre별 평균 Views

```python
dfm.groupby("Genre")["Views"].mean()
```

11. Genre별 최대 Rating

```python
dfm.groupby("Genre")["Rating"].max()
```

12. Rating이 평균보다 높은 row

```python
# dfm["Rating"].mean()은 전체 row에서 Rating의 평균입니다.

cond = dfm["Rating"] > dfm["Rating"].mean()
dfm[cond]
```
